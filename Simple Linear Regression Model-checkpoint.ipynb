```python
# Importing relevant libraries

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# generating input data basing on which we will train the linear model "2x + 3z + 5"

observations = 1000

xs = np.random.uniform(-10,10,(observations,1))
zs = np.random.uniform(-10,10,(observations,1))


inputs = np.column_stack((xs,zs))
print(inputs)
```

```python
#Create the target equation which is like 2 * x - 3 * z + 5 + noise

noise = np.random.uniform(-1,1,(observations,1))

targets = 2 * xs - 3 * zs + 5 + noise
```

```python
# to have a visual representation of the target data to see the linear relationship
targets = targets.reshape(observations,)
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot(xs,zs,targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('targets')
ax.view_init(azim=100)
plt.show()
targets = targets.reshape(observations,1)
```

```python
# Building the objective function like y = w1 * x1 + w2 * x2 + b for example size and proximity of apartments 

init_range = 0.1

weights = np.random.uniform(-init_range,init_range,(2,1))

biases = np.random.uniform(-init_range,init_range,1)

print(weights,biases)
```

```python
learning_rate = 0.02
```

```python
for i in range(1000):
    outputs = np.dot(inputs,weights) + biases
    deltas = outputs - targets 
    
    loss = np.sum(deltas ** 2) / 2 / observations
    print(loss)
    
    deltas_scaled = deltas / observations
    
    weights = weights - learning_rate * np.dot(inputs.T,deltas_scaled)
    biases = biases - learning_rate * np.sum(deltas_scaled)
    
```

```python
print(weights,biases)
```

```python
plt.plot(outputs,targets)
plt.show()
```
